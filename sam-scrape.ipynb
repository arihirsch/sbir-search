```python
import requests
import json
import time
import pprint

API_URL = "https://api.sam.gov/prod/opportunities/v2/search"
API_KEY = ""
PARAMS = {
    "limit": 1000,  # Adjust as needed
    "offset": 0,
    "postedFrom": "01/01/2025",
    "postedTo": "02/11/2025",
    "ptype": "o",  # Open opportunities
    #"organizationName": "National Aeronautics and Space Administration",
    "api_key": API_KEY
}

all_opportunities = []

while True:
    response = requests.get(API_URL, params=PARAMS)
    if response.status_code != 200:
        print("Error fetching data:", response.text)
        break

    data = response.json()
    if not data.get("opportunitiesData"):
        break  # No more results

    all_opportunities.extend(data["opportunitiesData"])
    
    # Move to next page
    PARAMS["offset"] += PARAMS["limit"]

    # Respect API limits
    time.sleep(2)

# Save data
with open("funding_opportunities.json", "w") as f:
    json.dump(all_opportunities, f, indent=4)

print(f"Fetched {len(all_opportunities)} opportunities.")
```
